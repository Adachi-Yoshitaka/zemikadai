```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_lfw_people
from sklearn.datasets import load_digits
```

```python
digits=load_digits()

fig, axes = plt.subplots(2,5, figsize=(10,5), gridspec_kw={"hspace":.5}, subplot_kw={"xticks":(), "yticks":()})

for ax, img in zip(axes.ravel(),digits.images):
    ax.imshow(img)
```

```python
print(digits.data.shape)
```

```python
import matplotlib.cm as cm
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
digits_pca=pca.fit_transform(digits.data)

plt.figure(figsize=(10,10))

plt.scatter(digits_pca[:,0],digits_pca[:,1],c=digits.target,cmap=cm.tab10)
plt.colorbar()
plt.xlabel("pca1")
plt.ylabel("pca2")
plt.show
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
digits_tsne=tsne.fit_transform(digits.data)

plt.figure(figsize=(10,10))

plt.scatter(digits_tsne[:,0],digits_tsne[:,1],c=digits.target,cmap=cm.tab10)
plt.colorbar()
plt.xlabel("tsne1")
plt.ylabel("tsne2")
plt.show
```
